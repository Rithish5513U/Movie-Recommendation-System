```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

movies = pd.read_csv('data/tmdb_5000_movies.csv')
credits = pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movie_features = movies[['id','title','original_title','genres','keywords','overview','production_companies','tagline']]
movie_features.head()
```

```python
credits.head()
```

```python
credits = credits.rename(columns={'movie_id':'id'})
credits.head()
```

```python
merged_df = pd.merge(movie_features, credits, on=['id','title'])
merged_df.head()
```

```python
merged_df.drop(columns=['original_title'], inplace=True)
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.drop(columns=['tagline'], inplace=True)
merged_df.head()
```

```python
merged_df.dropna(inplace=True)
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.duplicated().sum()
```

```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'].lower())
    return l
```

```python
merged_df['genres'][0]
```

```python
merged_df['genres'] = merged_df['genres'].apply(convert)
merged_df.head()
```

```python
merged_df['keywords'][0]
```

```python
merged_df['keywords'] = merged_df['keywords'].apply(convert)
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df['production_companies'][0]
```

```python
merged_df['production_companies'] = merged_df['production_companies'].apply(convert)
merged_df.head()
```

```python
merged_df['cast'][0]
```

```python
def convert_1(text):
    l = []
    cnt=0
    for i in ast.literal_eval(text):
        if cnt==3:
            break
        l.append(i['name'].lower())
        cnt+=1
    return l

merged_df['cast'] = merged_df['cast'].apply(convert_1)
merged_df.head()
```

```python
def director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
merged_df['crew'] = merged_df['crew'].apply(director)
merged_df.head()
```

```python
def remSpace(text):
    l = []
    for i in text:
        l.append(i.replace(" ",""))
    return l

merged_df['overview'] = merged_df['overview'].apply(lambda x:x.split())

merged_df['production_companies'] = merged_df['production_companies'].apply(remSpace)
merged_df['keywords'] = merged_df['keywords'].apply(remSpace)
merged_df['cast'] = merged_df['cast'].apply(remSpace)
merged_df['crew'] = merged_df['crew'].apply(remSpace)
merged_df['genres'] = merged_df['genres'].apply(remSpace)
merged_df['overview'] = merged_df['overview'].apply(remSpace)

merged_df.head()
```

```python
merged_df['tag'] = merged_df['genres']+merged_df['keywords']+merged_df['overview']+merged_df['production_companies']+merged_df['cast']+merged_df['crew']
merged_df.head()
```

```python
merged_df.drop(columns=['genres','keywords','overview','production_companies','cast','crew'], inplace=True)
merged_df.head()
```

```python
merged_df['tag'] = merged_df['tag'].apply(lambda x:" ".join(x))
merged_df.head()
```

```python
merged_df['tag'][0]
```

```python
import nltk
from nltk.stem import PorterStemmer

pt = PorterStemmer()

def stems(text):
    l=[]
    for i in text.split():
        l.append(pt.stem(i))
    return " ".join(l)

merged_df['tag'] = merged_df['tag'].apply(stems)
merged_df.head()
```

```python
merged_df['tag'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

vector = cv.fit_transform(merged_df['tag']).toarray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = merged_df[merged_df['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    for i in distances[1:6]:
        print(merged_df.iloc[i[0]].title)

recommend('Spectre')
```

```python
import pickle as pk

pk.dump(merged_df, open('Artifacts/movie_list.pkl','wb'))
pk.dump(similarity, open('Artifacts/similarity.pkl','wb'))
```
